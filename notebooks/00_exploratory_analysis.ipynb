```python
df_empty_sources = df[df['Data_Sources'].isna()]
df_empty_sources.info()
df_empty_sources
```

Reviewing the entries with missing data sources, we find that they are centralized to a few indicators for select
cities. We also found that missing values persists for multiple years in these instances. Since we cannot verify the
sources, we will drop these entries in the preprocessing stage and supplement the data from reliable sources.

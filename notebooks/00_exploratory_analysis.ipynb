### Load and evaluate data

This project will look at the following datasets:
- cdc_places_county
- census_acs5_2023
- epa_aqs_annual_2023
- fred_series

Please visit https://ryanwaked.com/data for full information on the data

```python
# Connect to the database
con = duckdb.connect('md:us_gov_db')

datasets = {
    'cdc_places_county': 'raw.cdc_places_county',
    'census_acs5_2023': 'raw.census_acs5_2023',
    'epa_aqs_annual_2023': 'raw.epa_aqs_annual_2023',
    'fred_series': 'raw.fred_series',
}

dfs = {}

for name, table in datasets.items():
    path = f'../data/raw/{name}.csv'
    if os.path.exists(path):
        dfs[name] = pd.read_csv(path)
    else:
        df = con.sql(f"SELECT * FROM {table}").df()
        df.to_csv(path, index=False)
        dfs[name] = df

cdc_places_county_df = dfs['cdc_places_county']
census_acs5_2023_df  = dfs['census_acs5_2023']
epa_aqs_annual_2023_df = dfs['epa_aqs_annual_2023']
fred_series_df = dfs['fred_series']
```
